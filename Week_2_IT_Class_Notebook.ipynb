# Gaussian Distribution
The example below creates a Gaussian PDF with a sample space from -5 to 5, a mean of 0, and a standard deviation of 1. A Gaussian with these values for the mean and standard deviation is called the Standard Gaussian.

```python
# plot the gaussian pdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm
# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
#print(sample_space)
mean = 0.0
stdev = 1.0
#calculate the pdf
pdf = norm.pdf(sample_space, mean, stdev)
# plot
pyplot.plot(sample_space, pdf)
pyplot.show()
```

```python
# plot the gaussian cdf
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm
# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
# calculate the cdf
cdf = norm.cdf(sample_space)
# plot
pyplot.plot(sample_space, cdf)
pyplot.show()
```

# Law of Large Numbers
The law of large numbers helps us understand why we cannot trust a single observation from an
experiment in isolation. We expect that a single result or the mean result from a small sample
is likely. A result being likely means that it is close to the central tendency, the mean of the
population distribution. Similarly, an unlikely result is an observation or sample mean that is
far from the central tendency. The law reminds us to repeat the experiment in order to develop
a large and representative sample of observations before we start making inferences about what
the result means.

# An Example
We can demonstrate the law of large numbers with a small worked example. First, we can
design an idealized underlying distribution. We will use a Gaussian distribution with a mean of
50 and a standard deviation of 5. The expected value or mean of this population is therefore 50.
Below is some code that generates a plot of this idealized distribution.

```python
# idealized population distribution
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# x-axis for the plot
xaxis = arange(30, 70, 1)
print(xaxis)

# y-axis for the plot
yaxis = norm.pdf(xaxis, 50, 5)
print(yaxis)

# plot ideal population
pyplot.plot(xaxis, yaxis)
pyplot.show()
```

Now, we can pretend to forget everything that we know about the population and make
independent random samples from the population. We can create samples of different sizes and
calculate the mean. Given our intuition and the law of large numbers, we expect that as the
size of the sample is increased, the sample mean will better approximate the population mean.
The example below calculates samples of different sizes, estimates the mean and plots the errors
in the mean from the expected mean (50).

```python
# demonstrate the law of large numbers
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import array
from matplotlib import pyplot
# seed the random number generator
seed(1)
# sample sizes
sizes = list()
for x in range (10, 20000, 200):
    sizes.append(x)

print(sizes)

# generate samples of different sizes and calculate their means
means = [mean(5 * randn(size) + 50) for size in sizes]

listdata = [(5 * randn(10) + 50)]
print(listdata)

print(means)

# plot sample mean error vs sample size
pyplot.scatter(sizes, array(means)-50)
pyplot.show()
```

# Central Limit Theorem
The CLT theorem states that as the size of the sample increases, the distribution of the mean across
multiple samples will approximate a Gaussian distribution.

The central limit theorem is often confused with the law of large numbers by beginners. The
law of large numbers is another different theorem from statistics. It is simpler in that it states
that as the size of a sample is increased, the more accurate of an estimate the sample mean will
be of the population mean. The central limit theorem does not state anything about a single
sample mean; instead, it is broader and states something about the shape or the distribution of
sample means.

# An Example with Dice
We can make the central limit theorem concrete with a worked example involving the rolling of
die. Remember that a die is a cube with a different number on each side from 1-to-6. Each
number has a 1-in-6 likelihood to turn up from a roll. The distribution of the numbers that turn
up from a dice roll is uniform given the equal likelihood.

```python
# generate random dice rolls
from numpy.random import seed
from numpy.random import randint
from numpy import mean

# seed the random number generator
seed(1)

# generate a sample of dice rolls
rolls = randint(1, 7, 50)
print(rolls)
print(mean(rolls))
```

Running the example generates and prints the sample of 50 dice rolls and the mean value of
the sample. We know that the mean value of the distribution is 3.5 calculated as (1+2+3+4+5+6)/6 or
21/6 . We can see that the mean of the sample is slightly wrong, which is to be expected because
it is an estimate of the population mean.

This example above is the result of rolling the simulated dice 50 times. We can then repeat this process
multiple times, such as 1,000. This will give us a result of 1,000 sample means. According to the
central limit theorem, the distribution of these sample means will be Gaussian. The example
below performs this experiment and plots the resulting distribution of sample means.

```python
# demonstration of the central limit theorem
from numpy.random import seed
from numpy.random import randint
from numpy import mean
from matplotlib import pyplot
# seed the random number generator
seed(1)

# calculate the mean of 50 dice rolls 1000 times
means = [mean(randint(1, 7, 5000)) for _ in range(10000)]

#print(means)

# plot the distribution of sample means
pyplot.hist(means)
pyplot.show()
```

Running this example above creates a histogram plot of the sample means. We can tell from the
shape of the distribution that the distribution is Gaussian. Itâ€™s interesting to note the amount
of error in the sample mean that we can see in 1,000 trials of 50 dice rolls. Further, the central
limit theorem also states that as the size of each sample, in this case 50, is increased, then the
better the sample means will approximate a Gaussian distribution.

# Confidence Intervals for the Mean
If we know that the data come from normal distribution with the known sample mean and standard deviation, we can construct an interval around the mean. In this case, we are assuming the sample is large, so we can estimate the population variance.

Running the below example prints the confidence interval for the mean.

```python
from scipy import stats
import numpy as np

a = [10.3, 10.5, 9.6, 9.7, 10.6, 9.9, 10.1, 10.1, 9.9, 10.5,
    10.1, 10.1, 9.9, 9.8, 10.6, 10.0, 9.9, 10.0, 10.3, 10.1,
    10.1, 10.3, 10.5, 9.7, 10.1, 9.7, 9.8, 10.3, 10.2, 10.2,
    10.1, 10.5, 10.0, 10.0, 10.6, 10.9, 10.1, 10.1, 9.8, 10.7,
    10.3, 10.4, 10.4, 10.3, 10.4, 9.9, 9.9, 10.5, 10.0, 10.7,
    10.1, 10.6, 10.0, 10.7, 9.8, 10.4, 10.3, 10.0, 10.5, 10.1]

conf_int_percentage = 0.99

mean = np.mean(a)
sigma = np.std(a) # population std deviation

conf_int = stats.norm.interval(conf_int_percentage, loc=mean, scale=sigma/np.sqrt(len(a)))

print(mean)
print(conf_int)
```

# Confidence Intervals for the Variance
When the data come from a normal distribution (with a known sample mean), we can use confidence interval for the population mean. On the other hand, when the sample is not sufficiently large (most of the time in practice), the sample mean cannot be a good estimator of the population mean. Therefore, confidence interval for the mean would not be a good metric.

It is also often useful to find a confidence interval for the population variance. This is important, for example, in
quality control. However the distribution of the sample variance is not normal. To find a confidence
interval for the population variance we need to use another distribution called the "chi-squared"
distribution.

Running the below example prints the confidence interval for the variance and standard deviation.

```python
from scipy.stats import chi2
from statistics import variance

sample = [2.02, 1.94, 2.09, 1.95, 1.98, 2.00, 2.03, 2.04, 2.08, 2.07, 1.99, 1.96, 1.99, 1.95, 1.99, 1.99, 2.03, 2.05, 2.01, 2.03]
alpha = 0.05

# calculate tail percentages according to given alpha
left_tail = alpha/2 #0.025
right_tail = 1 - alpha/2 # 0.0975

# calculate degree of freedom
dof = len(sample)-1 #19

# calculate critical values of chi-squared distribution
X_left = chi2.ppf(1-left_tail, dof)
X_right = chi2.ppf(1-right_tail, dof)

left_bound_variance = (variance(sample)*dof)/X_left
right_bound_variance = (variance(sample)*dof)/X_right

print("left_bound: ", left_bound_variance**0.5)

print("right_bound: ", right_bound_variance**0.5)

```

We can then use these observations to make a claim about the sample distribution, such as:
- There is a 95% likelihood that the range 0.033 to 0.063 covers the variance.
